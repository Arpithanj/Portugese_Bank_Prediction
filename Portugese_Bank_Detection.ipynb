```python
## Handle Outliers By Box Plot Visualisation
data.loc[data['age']>90,'age'] = data.age.median()
data.loc[data['duration']>3000,'duration'] = data.duration.median()
data.loc[data['campaign'] > 35,'campaign'] = data.campaign.median()
#data.loc[data['cons.conf.idx'] >-30,'cons.conf.idx'] = data['cons.conf.idx'].median()
```

#### We replaced outliers (very high values) with the median for age, duration, and campaign columns to reduce noise and improve model performance.

 # ( iv )  SMOTE ( Data Balancing )

```python
# calculate the Target Class
data['y'].value_counts()

# we can Clearly see that the Target Class is Not Balanced , so we do Data Balance By SMOTE method
```

```python
# Create X & Y Variable for SMOTE
X = data.drop('y',axis=1)
Y = data['y']
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# Suppose X and Y are your features and target
# Identify and encode categorical columns
label_enc = LabelEncoder()

for col in X.select_dtypes(include=['object']).columns:
    X[col] = label_enc.fit_transform(X[col])

# Apply SMOTE
smote = SMOTE()
X_smot, y_smot = smote.fit_resample(X, Y)

print("Before SMOTE:", Y.value_counts())
print("After SMOTE:", y_smot.value_counts())
```

```python
y_smot.value_counts()
# Now Data class is Balanced
```

```python
# Create a data frame using Two SMOTE values
data = pd.concat([X_smot, y_smot],axis=1)
data
```

### Insights After Applying SMOTE (Data Balancing)

1 . Original Class Distribution Was Highly Imbalanced
Before applying SMOTE, the target variable y had:

39,922 samples for class 0 (did not subscribe)
5,289 samples for class 1 (subscribed)
This means class 1 represented only ~11.7% of the data, which can cause:
Biased models
Poor recall for positive class
High accuracy but misleading results

2.Why SMOTE Was Necessary
SMOTE generates synthetic samples for the minority class instead of duplicating them.
This helps:
Improve model’s ability to detect customers who subscribe
Reduce bias towards the majority class
Maintain feature-space patterns instead of creating exact duplicates

3.Balanced Distribution After SMOTE
After applying SMOTE:
Both class 0 and class 1 now have 39,922 samples
This creates a 50:50 balance, which is ideal for classification models.

4.Impact on Model Performance
SMOTE is expected to:
Increase Recall for class 1 (because the model now sees more positive samples)
Slightly reduce Precision, since synthetic samples may add noise
Provide a more reliable F1-score
Improve overall model generalization on minority class detection

5.Important Note
SMOTE should always be applied after train-test split to prevent data leakage.
(If you applied it before splitting, consider fixing it.)

## 6 . Model Creation & Evaluation

```python
# split the data into train & test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=15)

# import Required metrics to see the performance
from sklearn.metrics import accuracy_score,classification_report,f1_score,precision_score,recall_score
```

```python
# Models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
```

```python
#Step 2: Create Model Dictionary

# Create a dictionary of all models
models = {
    "Logistic Regression": LogisticRegression(),
    "K-Nearest Neighbour": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Support Vector Machine": SVC(kernel="linear"),   # faster kernel
    "Artificial Neural Network": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500),
    "XGBoost": XGBClassifier(eval_metric="logloss")
}
```

## Separate Evaluation for Each Model


```python
# 🔹 Logistic Regression

lr = models["Logistic Regression"]
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

#### This code trains a Logistic Regression model, makes predictions on the test data, and prints the accuracy and classification report. 
#### It helps you understand how well the simplest linear model performs compared to more advanced algorithms.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Base model
knn = KNeighborsClassifier()

# FAST parameter grid
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Fast GridSearchCV
grid_knn = GridSearchCV(knn, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_knn.fit(X_train, y_train)

# Best parameters
print("Best Hyperparameters for KNN:", grid_knn.best_params_)

# Best model
best_knn = grid_knn.best_estimator_

# Predictions
y_pred_knn = best_knn.predict(X_test)

# Evaluation
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

#### The KNN model achieved strong performance after tuning with GridSearchCV. 

#### The best hyperparameters selected were n_neighbors = 3, weights = distance, and metric = manhattan.

#### These optimized settings significantly improved prediction quality, resulting in an accuracy of 93.5% and consistently high precision, recall, and F1-scores for both classes.

#### Overall, the tuned KNN model demonstrates excellent ability to correctly classify both positive and negative cases.

```python
# 🔹 Decision Tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Simple Decision Tree
dt = DecisionTreeClassifier(random_state=42)

# SMALL parameter grid (FAST)
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# FAST GridSearch
grid_dt = GridSearchCV(dt, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_dt.fit(X_train, y_train)

# Best parameters
print("Best Hyperparameters:", grid_dt.best_params_)

# Best model
best_dt = grid_dt.best_estimator_

# Predict
y_pred_dt = best_dt.predict(X_test)

# Evaluation
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

#### The Decision Tree model performed strongly after tuning with GridSearchCV. 

#### The best hyperparameters identified were criterion='entropy', max_depth=None, min_samples_split=2, and min_samples_leaf=1, meaning the model works best as a fully grown tree. 

#### With an accuracy of 93.33%, the model shows balanced precision and recall (≈93% for both classes), indicating it can classify both classes effectively. 

#### Overall, the tuned Decision Tree provides high interpretability along with strong predictive performance.

```python
# 🔹 Random Forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Base model
rf = RandomForestClassifier(random_state=42)

# FAST parameter grid
param_grid = {
    'n_estimators': [50, 100, 150],     # small and fast
    'max_depth': [None, 5, 10],         # shallow + full
    'min_samples_split': [2, 5],        # basic splits
    'min_samples_leaf': [1, 2]          # small leaves
}

# Fast GridSearch
grid_rf = GridSearchCV(rf, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_rf.fit(X_train, y_train)

# Best hyperparameters
print("Best Hyperparameters for Random Forest:", grid_rf.best_params_)

# Best model
best_rf = grid_rf.best_estimator_

# Predictions
y_pred_rf = best_rf.predict(X_test)

# Evaluation
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

```

#### The tuned Random Forest model achieved a strong accuracy of 94.8%, outperforming the Decision Tree and providing more stable, generalized predictions. 

#### GridSearchCV identified the best setup as 150 trees, no depth limit, and minimal split/leaf settings, allowing the model to learn complex patterns while reducing overfitting through ensembling. Precision, recall, and F1-score are all around 95% for both classes, showing excellent balance and reliability. 

#### Overall, Random Forest proves to be a high-performing and robust model for this dataset.”

```python
# 🔹 Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Base model
gb = GradientBoostingClassifier(random_state=42)

# FAST parameter grid
param_grid = {
    'n_estimators': [50, 100, 150],      # small and quick
    'learning_rate': [0.01, 0.05, 0.1],  # most common
    'max_depth': [2, 3, 4],              # shallow trees work best for GB
    'subsample': [1.0, 0.8]              # reduces overfitting
}

# Fast GridSearch
grid_gb = GridSearchCV(
    gb, param_grid, cv=3, scoring='accuracy', n_jobs=-1
)
grid_gb.fit(X_train, y_train)

# Best Hyperparameters
print("Best Hyperparameters for Gradient Boosting:", grid_gb.best_params_)

# Best Model
best_gb = grid_gb.best_estimator_

# Predictions
y_pred_gb = best_gb.predict(X_test)

# Evaluation
print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
print(classification_report(y_test, y_pred_gb))


```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Base SVM model
svm = SVC(random_state=42)

# FAST parameter grid (very small)
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']   # used only for rbf, but safe to include
}

# Fast GridSearch
grid_svm = GridSearchCV(
    svm, param_grid,
    cv=3,           # fast 3-fold CV
    scoring='accuracy',
    n_jobs=-1       # use all CPU cores
)

# Fit model
grid_svm.fit(X_train, y_train)

# Best params
print("Best Hyperparameters for SVM:", grid_svm.best_params_)

# Best estimator
best_svm = grid_svm.best_estimator_

# Predictions
y_pred_svm = best_svm.predict(X_test)

# Evaluation
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

#### Hyperparameter tuning improved the SVM performance.

#### GridSearchCV found the best combination:

#### C = 10 → gives stronger decision boundary

#### kernel = rbf → captures non-linear patterns

#### gamma = auto → adapts based on dataset size

#### This tuned model increased accuracy to ~93%, making SVM perform better than the default version.

```python
# ------------------ ANN  ------------------

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# Scale data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build ANN
ann = Sequential([
    Dense(32, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile ANN
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train ANN
ann.fit(X_train_scaled, y_train, epochs=20, batch_size=32, verbose=1)

# Predictions
y_pred_ann = ann.predict(X_test_scaled)
y_pred_ann = (y_pred_ann > 0.5).astype(int)

# Metrics
print("ANN Accuracy:", accuracy_score(y_test, y_pred_ann))
print(classification_report(y_test, y_pred_ann))

# ---------------------------------------------------------------
```

#### The model used Adam optimizer for faster convergence and performed well after scaling the data.

#### After training for 20 epochs, the ANN achieved an accuracy of ~92%, showing that it can learn complex relationships in the dataset but performs slightly lower than tuned tree-based models like Random Forest.

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Base model
xgb = XGBClassifier(
    eval_metric='logloss',
    random_state=42,
    use_label_encoder=False
)

# FULL BUT FAST parameter grid
param_grid = {
    "n_estimators": [100, 200, 300],     # number of trees
    "learning_rate": [0.01, 0.05, 0.1], # speed of learning
    "max_depth": [3, 5, 7],             # tree depth
    "subsample": [0.7, 0.8, 1.0],       # % of rows used
    "colsample_bytree": [0.7, 0.8, 1.0] # % of features used
}

# GridSearch
grid_xgb = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring="accuracy",
    cv=3,
    n_jobs=-1
)

# Fit
grid_xgb.fit(X_train, y_train)

# Best parameters
print("Best Hyperparameters for XGBoost:", grid_xgb.best_params_)

# Best model
best_xgb = grid_xgb.best_estimator_

# Predict
y_pred_xgb = best_xgb.predict(X_test)

# Results
print("Tuned XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

#### XGBoost was tuned using a focused GridSearch to improve accuracy by adjusting tree depth, learning rate, number of trees, and sampling ratios.

#### The best model used 300 trees, learning rate 0.1, max depth 7, and balanced sampling of rows and features.

#### This tuning significantly improved model performance, achieving ~94.5% accuracy, making XGBoost one of the strongest models in your project with excellent precision and recall for both classes.

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

combined_results = {
    "Model": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1 Score": [],
    "Best Hyperparameters": []
}

def add_combined(model_name, y_test, y_pred, best_params):
    combined_results["Model"].append(model_name)
    combined_results["Accuracy"].append(round(accuracy_score(y_test, y_pred), 4))
    combined_results["Precision"].append(round(precision_score(y_test, y_pred, average='weighted'), 4))
    combined_results["Recall"].append(round(recall_score(y_test, y_pred, average='weighted'), 4))
    combined_results["F1 Score"].append(round(f1_score(y_test, y_pred, average='weighted'), 4))
    combined_results["Best Hyperparameters"].append(best_params)

# Add all models (replace variable names with yours)
add_combined("Logistic Regression", y_test, y_pred_lr,"Not Tuned")
add_combined("KNN", y_test, y_pred_knn, grid_knn.best_params_)
add_combined("Decision Tree", y_test, y_pred_dt, grid_dt.best_params_)
add_combined("Random Forest", y_test, y_pred_rf, grid_rf.best_params_)
add_combined("Gradient Boosting", y_test, y_pred_gb, grid_gb.best_params_)
add_combined("XGBoost", y_test, y_pred_xgb, grid_xgb.best_params_)
add_combined("SVM", y_test, y_pred_svm, grid_svm.best_params_)
add_combined("ANN", y_test, y_pred_ann,"Not Tuned")

# Convert to DataFrame
combined_df = pd.DataFrame(combined_results)

print("\nFINAL MODEL + HYPERPARAMETER COMPARISON TABLE:")
print(combined_df)
```

### Challenges Faced During the Project & Techniques Used

Imbalanced Target (y)

Most clients said “no”, very few said “yes”.
-Used SMOTE / class weights to handle imbalance.

Many Categorical Variables
(job, marital, education, contact, month, poutcome)
-Applied One-Hot Encoding to convert them for ML models.

Outdated / Rare Categories
(e.g., unknown jobs)
-Grouped rare categories to avoid model noise.

Multicollinearity
-Checked correlations; dropped redundant features.

Scaling Required
-StandardScaler used for KNN, SVM, ANN.

Overfitting in Tree Models
-Controlled using max_depth, min_samples_split, subsample.

Hyperparameter Tuning Was Slow
-Used optimized small grids for faster, efficient tuning.

### Final Recommendation

Deploy the Random Forest model for production since it achieves the highest accuracy, balanced precision–recall, strong generalization, and minimal tuning complexity.

### Why NOT the others

KNN: High accuracy but slow for large datasets (distance-based).

Decision Tree: Overfits easily compared to Random Forest.

Gradient Boosting / XGBoost: Excellent but slightly lower accuracy in your results.

SVM: Good but slower and harder to tune.

ANN: Lower accuracy and more complex for deployment.

Logistic Regression: Baseline model only.
